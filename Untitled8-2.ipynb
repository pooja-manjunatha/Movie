```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
uri = "bolt://localhost:7687"
driver = GraphDatabase.driver(uri, auth=("neo4j", "password"))
```

```python
query = """
MATCH (m:Movie)
WHERE m.imdb_score IS NOT NULL
RETURN m.title AS title, m.imdb_score AS imdb_score
ORDER BY imdb_score DESC
LIMIT 5
"""

with driver.session() as session:
    result = session.run(query)
    data = [(record["title"], record["imdb_score"]) for record in result]
colors = ['red', 'green', 'blue', 'orange','yellow']
plt.figure(figsize=(10,6))
plt.bar([x[0] for x in data], [x[1] for x in data], color=colors)
plt.xlabel("Movie Title")
plt.ylabel("IMDB Rating")
plt.title("Top 5 Highest Rated Movies")
plt.xticks(rotation=45)

for i, v in enumerate([x[1] for x in data]):
    plt.text(i-0.2, v+0.1, str(v), color='white', fontweight='bold')

plt.show()
print("It can be seen in this plot that the highest rated movie is THE SHAWSHANK REDEMPTION followed by THE GODFATHER PART 1 AND 2, THE DARK KNIGHT AND THE RETURN OF THE KING")
```

```python
query = """
MATCH (m:Movie)
RETURN m.title_year AS release
"""

with driver.session() as session:
    result = session.run(query)
    data = pd.DataFrame([dict(record) for record in result])

sns.set(style="darkgrid")
sns.histplot(data=data, x="release", bins=30, kde=True)
plt.title("Histogram of Release years")
plt.xlabel("Release Year")
plt.ylabel("Frequency")
plt.show()
print("The histogram represents the distribution of movies released over the years. We can see that the number of movies has significantly increased since the year 2000, which is clearly shown by the steep rise in the number of movies towards the right side of the graph. This indicates that the film industry has been growing exponentially in the recent past")

```

```python
query = """
MATCH (m:Movie)
WHERE m.num_user_for_reviews IS NOT NULL AND m.imdb_score IS NOT NULL
RETURN m.num_user_for_reviews AS user_reviews, m.imdb_score AS imdb_score
"""
with driver.session() as session:
    result = session.run(query)
    data = pd.DataFrame(result.data())
sns.set(style="whitegrid")
sns.scatterplot(x="user_reviews", y="imdb_score", data=data)
plt.show()

print('A correlation coefficient of 0.29 indicates a weak positive correlation between IMDb score and number of user reviews. This means that as the IMDb score increases, the number of user reviews also tends to increase, but not necessarily by a large amount. The scatter plot being more concentrated on the left side from top to bottom could suggest that there are more movies with lower IMDb scores and fewer user reviews compared to movies with higher IMDb scores and more user reviews.')
corr_coeff = df['imdb_score'].corr(df['num_user_for_reviews'])

print(f"Correlation coefficient: {corr_coeff:.2f}")





```
